Find the stats associated with mctaco dataset

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize_text(text):
    return tokenizer.tokenize(text)

def count_sentences(text):
    return len(nltk.sent_tokenize(text))

token_lengths = []
sentence_counts = []

for split in ['test', 'validation']:
    for item in dataset[split]:
        context = item['sentence']
        tokens = context.split()
        num_sentences = count_sentences(context)
        # if num_sentences==2:
        #   print(context)
        token_lengths.append(len(tokens))
        # if len(tokens)==32:
        #   print(context)
        sentence_counts.append(num_sentences)

avg_tokens = np.mean(token_lengths)
max_tokens = np.max(token_lengths)
min_tokens = np.min(token_lengths)
avg_sentences = np.mean(sentence_counts)

print(f"Average number of tokens in context: {avg_tokens}")
print(f"Maximum number of tokens in context: {max_tokens}")
print(f"Minimum number of tokens in context: {min_tokens}")
print(f"Average number of sentences in context: {avg_sentences}")
```

Find the stats associated with RealNews dataset

```python
import os
import pandas as pd

#the folder contains 4 csv files which together constitute the training data
folder_path = '/content/drive/MyDrive/RealNews'

csv_list = []

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        csv_list.append(pd.read_csv(file_path))

combined_df = pd.concat(csv_list, ignore_index=True)

token_lengths = []
sentence_counts = []
for index, row in combined_df.iterrows():
  context=row['context']
  tokens = context.split()
  num_sentences = count_sentences(context)
  token_lengths.append(len(tokens))
  sentence_counts.append(num_sentences)

avg_tokens = np.mean(token_lengths)
max_tokens = np.max(token_lengths)
min_tokens = np.min(token_lengths)
avg_sentences = np.mean(sentence_counts)

print(f"Average number of tokens in context: {avg_tokens}")
print(f"Maximum number of tokens in context: {max_tokens}")
print(f"Minimum number of tokens in context: {min_tokens}")
print(f"Average number of sentences in context: {avg_sentences}")
```
